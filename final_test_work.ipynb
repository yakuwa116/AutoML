```python
ohe_columns = [ 'sales',
               'salary']

my_dtype = {'sales':object,
           'salary':object}

import pandas as pd
pd.options.display.max_columns = 50
```

```python
df = pd.read_csv('./final_hr_analysis_train.csv', header=0)

X = df.iloc[:, 2:]
y = df.iloc[:, 1]
```

```python
X_ohe = pd.get_dummies(X,
                      dummy_na=True,
                      columns=ohe_columns)

display(X_ohe.head())
```

```python
X_ohe.describe()
```

```python
from sklearn.impute import SimpleImputer

imp = SimpleImputer()
imp.fit(X_ohe)

X_ohe_columns = X_ohe.columns.values
X_ohe = pd.DataFrame(imp.transform(X_ohe),columns=X_ohe_columns)

display(X_ohe.head())
```

```python
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier

selector = RFE(estimator=RandomForestClassifier(n_estimators=100,random_state=0),
               n_features_to_select=10,
               step=.05)
selector.fit(X_ohe,y)

print('Done normally')
```

```python
print(selector.support_)
```

```python
X_fin = X_ohe.loc[:, X_ohe_columns[selector.support_]]

print('X_fin shape:(%i,%i)' % X_fin.shape)
X_fin.head()
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(X_fin,
                                                   y,
                                                   test_size=0.3,
                                                   random_state=0)

pipelines = {
    'knn':
        Pipeline([('scl', StandardScaler()),
                 ('est', KNeighborsClassifier())]),
    'logistic':
        Pipeline([('scl', StandardScaler()),
                 ('est', LogisticRegression(random_state=0))]),
    'svc':
        Pipeline([('scl', StandardScaler()),
                 ('est', SVC(C=10, kernel='rbf', class_weight='balanced', random_state=0))]),
    'lsvc':
        Pipeline([('scl', StandardScaler()),
                 ('est', LinearSVC(C=10, class_weight='balanced', random_state=0))]),
    'tree':
        Pipeline([('scl', StandardScaler()),
                 ('est', DecisionTreeClassifier(random_state=0))]),
    'rf':
        Pipeline([('scl', StandardScaler()),
                 ('est', RandomForestClassifier(random_state=0))]),
    'gb':
        Pipeline([('scl', StandardScaler()),
                 ('est', GradientBoostingClassifier(random_state=0))]),
}

scores = {}
for pipe_name, pipeline in pipelines.items():
    pipeline.fit(X_train, y_train)
    scores[(pipe_name, 'train')] = accuracy_score(y_train, pipeline.predict(X_train))
    scores[(pipe_name, 'test')] = accuracy_score(y_test, pipeline.predict(X_test))
pd.Series(scores).unstack()
```
